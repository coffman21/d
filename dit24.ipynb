```python
def count(df, column, sort=False):
    _ = df.groupby([column])[[column]].count()
    _.index.names = ['index']
    if sort:
        _ = _.sort_values([column])
    return _
```

```python
t = dit[dit['SUCCESS'] == True]
tc = count(t, 'ipAddr').rename(columns={'ipAddr':'succeed'})
f = dit[dit['SUCCESS'] == False]
fc = count(f, 'ipAddr').rename(columns={'ipAddr':'failed'})
res = pd.concat([tc, fc], axis=1).fillna(0)
```

```python
res.sort_values(['succeed', 'failed']).plot(stacked=True, figsize=(15,5), color=['b', 'r'])
plt.yscale('log')
plt.xticks([])
plt.show()

n = res.sort_values(['failed', 'succeed'])
n=n.reindex(columns=['failed', 'succeed'])
n.plot(stacked=True, figsize=(15,5), color=['r', 'b'])
plt.yscale('log')
plt.xticks([])
plt.show()
```

```python
def convert_ip(address):
    o1, o2, o3, o4 = [int(_) for _ in address.split('.')]
    integer_ip = 16777217*o1 + 65536*o2 + 256*o3 + o4 
    return integer_ip
```
